```python
def CVD(heart_1, heart_2, heart_3, heart_4, heart_5):
    if ((heart_1 == 1.0) or (heart_2 == 1.0) or (heart_3 == 1.0)
       or (heart_4 == 1.0) or (heart_5 == 1.0)):
        return 1
    else:
        return 0
```

```python
df['CVD_risk'] = df.apply(lambda x: CVD(x['congestive_fail'], x['coronary_disease'],
                                         x['angina'], x['heart_attack'], 
                                         x['stroke']), axis=1)
df = df.drop(['congestive_fail', 'coronary_disease', 'angina', 'heart_attack', 'stroke'], 1)
```

```python
def barplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.3, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        barlist=ax.bar(x = 'CVD_risk', height = col, data=df_grouped)
        barlist[1].set_color('sandybrown')
        ax.set_xticks([0, 1])
        ax.set(xlabel='CVD risk', ylabel=col)
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
def boxplots(columns, ncol=None, figsize=(15, 8)):
    if ncol is None:
        ncol = len(columns)
    nrow = int(np.ceil(len(columns) / ncol))
    fig, axes = plt.subplots(nrow, ncol, figsize=figsize, squeeze=False)
    fig.subplots_adjust(wspace=0.3, hspace=0.6)
    for i, col in enumerate(columns):
        ax = axes.flatten()[i]
        barlist=ax.boxplot(x=col, data=df)
        ax.set(xlabel=col)
    nsubplots = nrow * ncol    
    for empty in range(i+1, nsubplots):
        axes.flatten()[empty].set_visible(False)
```

```python
def income_cat(income):
    if ((income == 1) or (income == 2)
       or (income == 3) or (income == 4)
       or (income == 5)):
        return 'lowest'
    elif ((income == 6) or (income == 7)
          or (income == 8) or (income == 9)
         or (income == 10)):
        return 'lower middle'
    else:
        return 'middle & over'
```

```python
df['income_cat'] = df.apply(lambda x : income_cat(x['income']), axis=1)
```
